```python
#Importing the file
import pandas as pd
df = pd.read_csv("blogs.csv")
df
```

```python
df['Data,Labels'] = df['Data'] + df['Labels']
df
```

```python
df.drop(df.columns[[0,1]],axis=1,inplace=True)
df
```

```python
# Specify the path to your .txt file
file_path = 'blogs.csv'


```

```python
# Open and read the contents of the file
with open(file_path, 'r', encoding='utf-8') as file:
    text_data = file.read()
```

```python
#Preprocess the data by cleaning the text (removing punctuation, converting to lowercase, etc.), tokenizing, and removing stopwords.
# Tokenize the text
from nltk.tokenize import word_tokenize

words = word_tokenize(text_data)
words
```

```python
# Convert all words to lowercase
words2 = [i.lower() for i in words]
words2
```

```python
from nltk.corpus import stopwords
stopwords_list = stopwords.words('english')
stopwords_list
```

```python
# Remove stop words
#words3 = [word for word in words2 if word not in stopwords_list]
#words3

words3 = []

for word in words2:
    if word not in stopwords_list:
        words3.append(word)

words3

print("after stopwords:", len(words3))
```

```python
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~.....''``'s'''

words_without_punctuation = []

for word in words3:
    if word not in punctuations:
        words_without_punctuation.append(word)


print("removing the punctuations: ",len(words_without_punctuation))
```

```python
from nltk.stem import PorterStemmer

porter_stemmer = PorterStemmer()

stemmed_words = [porter_stemmer.stem(word) for word in words_without_punctuation]

print(len(stemmed_words))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a dictionary to store the word frequency
word_counts = {}

for word in stemmed_words:
  if word not in word_counts:
    word_counts[word] = 0
  word_counts[word] += 1


# Create the word cloud
wordcloud = WordCloud(width=800, height=600, background_color="white").generate_from_frequencies(word_counts)

# Display the word cloud
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
doc = " ".join(words4)
doc
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Create a CountVectorizer object with ngram_range=(2, 2) to extract bigrams
vect = CountVectorizer(ngram_range=(2, 2))

# Fit the vectorizer to the document
counts = vect.fit_transform([doc])
counts
```

```python
# Get the vocabulary of the vectorizer
vocab = vect.get_feature_names_out()
vocab
```

```python
# Get the top 20 bigram counts
top_20_bigrams = counts.toarray().sum(axis=0).argsort()[-20:]

# Create a bar chart of the top 20 bigram counts
plt.figure(figsize=(15, 7))
plt.bar(vocab[top_20_bigrams], counts.toarray()[0, top_20_bigrams])
plt.xticks(rotation=90)
plt.xlabel("Bigrams")
plt.ylabel("Count")
plt.title("Top 20 Bigram Counts")
plt.show()
```

```python
# prompt: read  every line from above df_final and convert in to either positive or negative statment and save under a new column

import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
analyzer = SentimentIntensityAnalyzer()
```

```python
# Create a new column for sentiment
df['Sentiment_0'] = df['Data,Labels'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
df.head()
```

```python
# Classify each statement as positive or negative based on the sentiment score
df['Sentiment'] = df['Sentiment_0'].apply(lambda x: 'positive' if x > 0 else 'negative')
df
```

```python
df["Sentiment"].value_counts()
```

```python
df['Data,Labels'] = df['Data,Labels'].str.lower()
df.head()
```

```python
import string
df['Data,Labels'] = df['Data,Labels'].str.replace('[{}]'.format(string.punctuation), '')
df.head()
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()


df['Data,Labels'] = df['Data,Labels'].apply(lambda x: " ".join([lemmatizer.lemmatize(word) for word in x.split()]))
df.head()
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

df['Data,Labels'] = df['Data,Labels'].apply(lambda x: " ".join([word for word in x.split() if word not in stop_words]))
df.head()
```

```python
Y = df["Sentiment"]
x = df['Data,Labels']
```

```python
# Caluculatinb TFIDF
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
#vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(x)
x_new = X.toarray()
x_new = pd.DataFrame(x_new)
x_new.to_csv("newdata.csv")
x_new
```

```python
# Data partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20)
```

```python
#fitting the naive bayes model
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train,Y_train)
```

```python
Y_pred_train=model.predict(X_train)
Y_pred_test=model.predict(X_test)
```

```python
#metrics
from sklearn.metrics import accuracy_score
ac1=accuracy_score(Y_train,Y_pred_train)
print("Training accuracy:",ac1.round(3))
ac2=accuracy_score(Y_test,Y_pred_test)
print("Test accuracy:",ac2.round(3))
```

```python
#Evaluate the performance of your Naive Bayes classifier using metrics such as accuracy, precision, recall, and F1-score.
```

```python
# Calculate accuracy
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report
accuracy = accuracy_score(Y_test,Y_pred_test)

# Calculate precision, recall, and F1-score for each class
precision = precision_score(Y_test,Y_pred_test,average='weighted')
recall = recall_score(Y_test,Y_pred_test, average='weighted')
f1 = f1_score(Y_test,Y_pred_test, average='weighted')

# Print the evaluation metrics
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")

# Detailed classification report
print("\nClassification Report:\n", classification_report(Y_test,Y_pred_test))
```
